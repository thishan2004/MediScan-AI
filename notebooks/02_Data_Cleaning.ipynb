```python
import pandas as pd
df = pd.read_csv("../data/raw/DDI_data.csv")
df.shape
```

```python
duplicates = df[df.duplicated()]
duplicates.head()
```

```python
df = df.drop_duplicates()
```

```python
df["drug1_name"] = df["drug1_name"].str.strip()
df["drug2_name"] = df["drug2_name"].str.strip()
df["interaction_type"] = df["interaction_type"].str.strip()
```

```python
df.to_csv("../data/processed/cleaned_DDI_data.csv", index=False)
```
